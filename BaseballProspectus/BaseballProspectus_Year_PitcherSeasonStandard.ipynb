```python
#要注意的地方有兩個 range, year, fileName
import string
import time
import requests
from bs4 import BeautifulSoup
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.93 Safari/537.36'
}
url='http://www.baseballprospectus.com/sortable/index.php'      
for year in range(1871,1950): #2015沒算到*************
    payload={                       
            'mystatslist':'LVL,NAME,YEAR,AGE,G,GS,PITCHES,IP,FAIR_IP,IP_START,IP_RELIEF,W,L,SV,BS,PA,AB,R,ER,H,B1,B2,B3,HR,TB,BB,UBB,IBB,SO,HBP,SF,SH,ERA,FAIR_RA,PPF,VORP,FIP,PVORP,PWARP',
            'category':'pitcher_season',
            'tablename':'dyna_pitcher_season',
            'stage':'data',
            'year':'%d'%(year),#**************
            'mystatslist':'LVL,NAME,YEAR,AGE,G,GS,PITCHES,IP,FAIR_IP,IP_START,IP_RELIEF,W,L,SV,BS,PA,AB,R,ER,H,B1,B2,B3,HR,TB,BB,UBB,IBB,SO,HBP,SF,SH,ERA,FAIR_RA,PPF,VORP,FIP,PVORP,PWARP',            
            'category':'pitcher_season',
            'tablename':'dyna_pitcher_season',
            'stage':'data',
            'group_LVL':'MLB',
            'minimum':'0',
            'sort1column':'NAME',
            'sort1order':'ASC',
            'page_limit':'1500',
            'glossary_terms':'*',
            'tt_team':'*',
            'show_ttroster':'1',
            'show_ttwatched':'1',
            'viewdata':'View Data',
            'start_num':'0',
        }
    res=requests.post(url,data=payload,headers=headers)
    soup=BeautifulSoup(res.text.encode('utf-8'))    
    trList=soup.select('#TTdata')[0].select('tr')
    fileName= str(year)+'_PitcherSeasonStandard.csv'
    fid=open('Baseball\\'+fileName,'w')
    for tr in trList:
        for td in tr.select('td'):
            fid.write(td.text.encode('utf-8')+',')
        fid.write('\n')
    fid.close()
    fod=open('Baseball\\'+fileName,'r')
    print 'year'+str(year)
    print fod.readlines()[1]    
    fod.close()
    time.sleep(2)
```
