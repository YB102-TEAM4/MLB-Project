```python
#首先Name要排序
#要注意的地方有四個 range, year, fileName, page_limit
import string
import time
import requests
from bs4 import BeautifulSoup
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.93 Safari/537.36'
}
url='http://www.baseballprospectus.com/sortable/index.php'      
for year in range(1871,2015): #2015沒算到**************
    payload={
            'mystatslist':'TEAM,LG,YEAR,G,PA,AB,R,H,HR,TB,BB,IBB,SO,BBR,SOR,HBP,SF,SH,RBI,SB,CS,SB_PERCENT,AVG,OBP,SLG,ISO,G_NUM',
            'category':'team_batting',
            'tablename':'dyna_team_batting',
            'stage':'data',
            'year':'%d'%(year),#*********************
            'mystatslist':'TEAM,LG,YEAR,G,PA,AB,R,H,HR,TB,BB,IBB,SO,BBR,SOR,HBP,SF,SH,RBI,SB,CS,SB_PERCENT,AVG,OBP,SLG,ISO,G_NUM',
            'category':'team_batting',
            'tablename':'dyna_team_batting',
            'stage':'data',
            'group_LG':'*',
            'minimum':'0',
            'sort1column':'G_NUM',
            'sort1order':'DESC',
            'sort2column':'TEAM',
            'sort2order':'ASC',
            'sort3column':'TEAM',
            'sort3order':'ASC',
            'page_limit':'All',#*********************
            'viewdata':'View Data',
        }
    res=requests.post(url,data=payload,headers=headers)
    soup=BeautifulSoup(res.text.encode('utf-8'))    
    trList=soup.select('#TTdata tr')
    fileName= str(year)+'_TeamBattingGuillenNumber.csv'#*********************
    fid=open('Baseball\\'+fileName,'w')    
    for tr in trList:
        for td in tr.select('td'):
            fid.write(td.text.encode('utf-8')+',')
        fid.write('\n')
    fid.close()
    fod=open('Baseball\\'+fileName,'r')
    print 'year'+str(year)
    print fod.readlines()[1]
    fod.close()
    time.sleep(2)
```
