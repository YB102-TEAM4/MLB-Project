```python
#首先Name要排序
#要注意的地方有四個 folder, range, year, page_limit
import string, time, requests, os
from bs4 import BeautifulSoup
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.93 Safari/537.36'
}
url='http://www.baseballprospectus.com/sortable/index.php'

folder='PitcherSeasonRates'#資料夾名稱*********************

path='Baseball\\'+folder #路徑名稱
if os.path.exists(path):
    print 'folder exists'
else:    
    os.mkdir(path)
    print 'folder not exists'  
    
for year in range(1871,2015): #2015沒算到**************
    payload={
            'mystatslist':'LVL,NAME,YEAR,AGE,G,GS,PITCHES,IP,IP_START,IP_RELIEF,BB9,SO9,HR9,UBBR,SOR,SO_BB,BABIP,WHIP,AVG,OBP,SLG,FAIR_RA,ERA,VORP,FIP',
            'category':'pitcher_season',
            'tablename':'dyna_pitcher_season',
            'stage':'data',
            'year':'%d'%(year),#*********************
            'mystatslist':'NAME,YEAR,AGE,G,GS,PITCHES,IP,IP_START,IP_RELIEF,BB9,SO9,HR9,UBBR,SOR,SO_BB,BABIP,WHIP,AVG,OBP,SLG,FAIR_RA,ERA,VORP,FIP',
            'submit':'Submit Statistics Selection',
            'stage':'data',
            'category':'pitcher_season',
            'groupcolumn':'',
            'filtercolumn':'IP',
            'required':'YEAR,NAME',
            'sort1column':'NAME',
            'sort1order':'ASC',
            'page_limit':'All',#*********************
        }
    res=requests.post(url,data=payload,headers=headers)
    soup=BeautifulSoup(res.text.encode('utf-8'))
    trList=soup.select('#TTdata tr')
    fileName= str(year)+'_'+folder+'.csv'
    fid=open(path+'\\'+fileName,'w')
    for tr in trList:
        for td in tr.select('td'):
            fid.write(td.text.encode('utf-8')+',')
        fid.write('\n')
    fid.close()
    fod=open(path+'\\'+fileName,'r')
    print 'year'+str(year)
    print fod.readlines()[1]
    fod.close()
    time.sleep(2)
```
