```python
#要注意的地方有四個 range, year, fileName, page_limit BatterSeasonStandard
import string
import time
import requests
from bs4 import BeautifulSoup
headers={
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.93 Safari/537.36'
}
url='http://www.baseballprospectus.com/sortable/index.php'      
for year in range(1871,2015): #2015沒算到**************
    payload={            
            'mystatslist':'NAME,YEAR,AGE,G,PA,AB,R,H,B1,B2,B3,HR,TB,BB,IBB,SO,HBP,SF,SH,RBI,OBI_PERCENT,DP,NETDP,SB,CS,SB_PERCENT,AVG,OBP,SLG,OPS,ISO,BPF,OPP_QUAL_OPS,EQA,VORP,FRAA,BWARP',
            'category':'batter_season',
            'tablename':'dyna_batter_season',
            'stage':'data',
            'year':'%d'%(year),#**************
            'mystatslist':'NAME,YEAR,AGE,G,PA,AB,R,H,B1,B2,B3,HR,TB,BB,IBB,SO,HBP,SF,SH,RBI,OBI_PERCENT,DP,NETDP,SB,CS,SB_PERCENT,AVG,OBP,SLG,OPS,ISO,BPF,OPP_QUAL_OPS,EQA,VORP,FRAA,BWARP',
            'category':'batter_season',
            'tablename':'dyna_batter_season',
            'stage':'data',
            'group_LVL':'MLB',
            'group_PRIMARY_POS_LABEL':'*',
            'minimum':'0',
            'sort1column':'NAME',
            'sort1order':'ASC',
            'page_limit':'All',
            'tt_team':'*',
            'show_ttroster':'1',
            'show_ttwatched':'1',
            'viewdata':'View Data',
        }
    res=requests.post(url,data=payload,headers=headers)
    soup=BeautifulSoup(res.text.encode('utf-8'))    
    trList=soup.select('#TTdata')[0].select('tr')
    fileName= str(year)+'_BatterSeasonStandard.csv'
    fid=open('Baseball\\'+fileName,'w')
    for tr in trList:
        for td in tr.select('td'):
            fid.write(td.text.encode('utf-8')+',')
        fid.write('\n')
    fid.close()    
    fod=open('Baseball\\'+fileName,'r')
    print 'year'+str(year)
    print fod.readlines()[1]    
    fod.close()
    time.sleep(2)
```
